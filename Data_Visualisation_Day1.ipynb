```python
import numpy as np
import matplotlib as mlp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(40,50)
y = np.arange(50,60)
x,y
```

```python
plt.plot(x,y,'r*--',linestyle = 'dashed',linewidth = 2,marker = 5, markersize = 4,label = 'x-y function ') #plot graph wiht styles
plt.title("X-Y plot",color = 'g') 
plt.xlabel("X-axis",color = 'g')
plt.ylabel("Y-axis",color = 'g')
plt.grid('b')
plt.legend(loc = 'lower right')
```

```python
days = [1,2,3,4,5,6,7]    # days of the week
Sales_1 = [160,150,140,145,175,165,180]  #Sales of comapny 1
Sales_2 = [70,90,160,150,140,145,175]    #Sales of company 2
```

```python
plt.plot(days,Sales_1,color ='r',linestyle = 'dashed',linewidth = 2,markersize = 3,label = 'Company_1 sales')
plt.plot(days,Sales_2,color = 'g',linestyle = 'dashed',linewidth = 3,markersize = 3.5 ,label = 'Company_2 sales')
plt.title("Plot of 2 companies over days of week",color = 'b')
plt.xlabel("Days of the week","m")
plt.ylabel("Sales of the week","c")
plt.grid()
plt.legend(loc = 'upper left')
```
